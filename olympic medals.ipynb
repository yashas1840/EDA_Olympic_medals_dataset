```python
corr_a=a.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr_a, square=True)
plt.show()
```

```python
global_sales_genre_platform = a.pivot_table(index='summer_participations', columns='winter_participations', values='total_participation', aggfunc=np.sum).dropna(axis=1,thresh=5)
global_sales_mean_genre_platform = a.pivot_table(index='summer_participations', columns='winter_participations', values='total_participation', aggfunc=np.mean).dropna(axis=1,thresh=5)
```

```python
plt.figure(figsize=(24,8))
plt.title("Heatmap of winter and_participations and total_participation")

sns.heatmap(data=global_sales_genre_platform, annot=True,fmt=".3g")
plt.show()
```

```python
plt.figure(figsize=(24,8))
plt.title("Heatmap of winter and_participations and total_participation")

sns.heatmap(data=global_sales_mean_genre_platform, annot=True, fmt=".3g")
plt.show()
```
